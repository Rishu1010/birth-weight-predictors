```python
#splitting dataset into x and y

y= df['bwt']  #dependent variable

```

```python
x= df.iloc[:,1:]
x.head()
```

```python
##train test split
## command : <pip install scikit-learn>

from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
lr = LinearRegression()
lasso = Lasso()
ridge = Ridge()
```

```python
lr.fit(x_train, y_train)
lasso.fit(x_train, y_train)
ridge.fit(x_train, y_train)
```

```python
lr_pred =lr.predict(x_test)
lasso_pred =lasso.predict(x_test)
ridge_pred = ridge.predict(x_test)  
```

```python
##model evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
lr_rscore = r2_score(y_test, lr_pred)
lasso_rscore = r2_score(y_test, lasso_pred)
ridge_rscore = r2_score(y_test, ridge_pred)
```

```python
lr_rscore = mean_squared_error(y_test, lr_pred)
lasso_rscore = mean_squared_error(y_test, lasso_pred)
ridge_rscore = mean_squared_error(y_test, ridge_pred)
```

```python
print("\nPerformance of Linear Regression model:")
print(f"R2 Score: {lr_rscore}")
print(f"MSE: ", mean_squared_error(y_test, lr_pred))

print("\nPerformance of Lasso Regression model:")
print(f"R2 Score: {lr_rscore}")
print(f"MSE: ", mean_squared_error(y_test, lasso_pred))


print("\nPerformance of Ridge Regression model:")
print(f"R2 Score: {lr_rscore}")
print(f"Mean Squared Error: {mean_squared_error(y_test, lr_pred)}")
```
